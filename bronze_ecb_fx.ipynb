```python
import requests
from pyspark.sql import Row

url = "https://data-api.ecb.europa.eu/service/data/EXR/D.USD.EUR.SP00.A?format=csvdata"
csv_text = requests.get(url).text

df_fx_raw = spark.createDataFrame(
    [Row(raw=csv_text)]
)

df_fx_raw.write.mode("overwrite").text(
    "Files/bronze/ecb/usd_eur_fx"
)



```
