```python
import requests
import json
from pyspark.sql import Row

url = "https://api.worldbank.org/v2/country/USA/indicator/NY.GDP.MKTP.CD?format=json&per_page=1000"
raw_json = requests.get(url).json()

df_raw_json = spark.createDataFrame(
    [Row(raw=json.dumps(raw_json))]
)

df_raw_json.write.mode("overwrite").json(
    "Files/bronze/worldbank/gdp_usa"
)



```
